```python
# Import necessary Libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
# Load th dataset and display the first few rows
job_data = pd.read_csv('/Users/dtsreddy/Desktop/sample.csv')
job_data.head()
```

```python
job_data = job_data.drop(columns=['name', 'id'])
x = job_data.drop(columns=['timesinHr'])
y = job_data['timesinHr']
```

```python
x.head()
print(x.shape)
```

```python
y.head()
```

```python
# create a ML model
model = LogisticRegression()
```

```python
# train the model
model.fit(x.values, y.values)
```

```python
# Predict using the trained model
predictions = model.predict([[8]])
print("predictions:", predictions)
```
